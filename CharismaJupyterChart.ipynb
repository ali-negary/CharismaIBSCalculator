```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
from collections import deque
import CharismaCollector
```

#### Do you want average of all (assets), average of (industry), or one (asset)?

```python
x_axis = []
y_axis = []
choice = input("Please enter one of the words in the parenthesis:\n")
# choice = 1
```

#### if you have chosen asset or industry, please enter the name of asset or industry:

```python
if choice != "all":
    name = input("Please enter name of asset or industry:\n")
else:
    name = 1
```

```python
def get_data():
    time, value = CharismaCollector.collector(choice, name)
    global x_axis
    global y_axis
    x_axis.append(time)
    y_axis.append(value)


fig = plt.figure(figsize=(9,5))
ax1 = fig.add_subplot(1, 1, 1)


def animate(i):
    get_data()
    ax1.clear()
    ax1.plot(x_axis, y_axis)
```

```python
ani = animation.FuncAnimation(fig, animate, interval=30000)
plt.show()
```
